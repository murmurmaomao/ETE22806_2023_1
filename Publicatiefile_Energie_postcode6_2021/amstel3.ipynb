```python
import pandas as pd
import numpy as np
import os
```

```python
df= pd.read_csv('Publicatiefile_Energie_postcode6_2021.csv')
df.replace('.', np.nan, inplace=True)
df.iloc[:,1:] = df.iloc[:,1:].astype(float)
```

```python
#Select the postcode containing 1011 and 1015
regex1 = '1101.*'
regex2 = '1105.*'
df = df[df.Postcode6.str.match(regex1)|df.Postcode6.str.match(regex2)]
```

```python
#Total current residential electricity consumption in the amstel III region
residential_electricity_consumption = df['Average_electricity supply_residences'].sum()*12
#Total current companies electricity consumption in the amstel III region
companies_electricity_consumption = df['Average_electricity supply_companies'].sum()*12

print('residential electricity consumption: {} KWh \n companies electricity consumption: {} KWh \n Amstel III electricity consumption: {} KWh'.format(residential_electricity_consumption,companies_electricity_consumption,residential_electricity_consumption+companies_electricity_consumption))
```
